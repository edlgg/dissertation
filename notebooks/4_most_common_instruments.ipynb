```python
def load_songs(path):
    songs = []
    for _, _, filenames in os.walk(path):
        for filename in filenames:
            try:
                song = mido.MidiFile(path + filename)
                print("Loading song: " + filename)
                songs.append(song)
            except:
                print("Error loading: " + filename + " will delete")
                os.remove(path + filename)
    print("Done loading songs")
    return  songs

songs = load_songs(jazz_large)
```

```python
def get_instrument_type(instrument):
    if instrument < 0:
        return "None"
    # if instrument < 1:
    #     return "0"
    if instrument < 9:
        return "Piano"
    if instrument < 17:
        return "Percussion"
    if instrument < 25:
        return "Organ"
    if instrument < 33:
        return "Guitar"
    if instrument < 41:
        return "Bass"
    if instrument < 49:
        return "Strings"
    if instrument < 57:
        return "Ensemble"
    if instrument < 65:
        return "Brass"
    if instrument < 73:
        return "Reed"
    if instrument < 81:
        return "Pipe"
    if instrument < 105:
        return "Synth"
    if instrument < 113:
        return "Ethnic"
    if instrument < 121:
        return "Percussive"
    if instrument < 129:
        return "Effect"
    return "Other"
```

```python
def get_n_most_common(song, n, class_="song"):
    #returns sorted list of n most common instruments

    tracks = song.tracks
    tracks.sort(key = len, reverse=True)
    # get list of instruments

    all_instruments = []
    for track in tracks:
        for msg in track:
            if msg.type == 'program_change' and len(track) > 50:
                if class_ == "type":
                    type = get_instrument_type(msg.program)
                    all_instruments.append(type)
                    # if msg.program == 0:
                    #     print(song)
                else:
                    all_instruments.append(msg.program)

    common_instruments = []
    for instrument in all_instruments:
        if instrument not in common_instruments:
            common_instruments.append(instrument)

    if len(common_instruments) < n:
        return [-1] * n

    common_instruments.sort()
    return common_instruments[:n]

love_me = mido.MidiFile('../../songs/small_db/Love_Me_Do.mid')
get_n_most_common(love_me, 2)
```

```python
def find_track_instrument_type(track):
    for msg in track:
        if msg.type == "program_change":
            return get_instrument_type(msg.program)
```

```python
def test_find_track_instrument_type():
    love_me = mido.MidiFile('../../songs/small_db/Love_Me_Do.mid')
    t = find_track_instrument_type(love_me.tracks[1])
    if t == "Organ":
        return True
    return False
test_find_track_instrument_type()
```

```python
def get_wanted_tracks(song, instrument_types):
    tracks = {}
    for track in song.tracks:
        t = find_track_instrument_type(track)
        # print(t)
        if t in instrument_types:
            if t not in tracks or len(track) > len(tracks[t]):
                tracks[t] = track
    wanted_tracks = []
    # print(tracks)
    for type_ in instrument_types:
        if type_ in tracks:
            wanted_tracks.append(tracks[type_])
        else:
            return [-1] * len(instrument_types)
    return wanted_tracks
```

```python
def test_get_wanted_tracks():
    love_me = mido.MidiFile('../../songs/small_db/Love_Me_Do.mid')
    types = ["Organ", "Bass"]
    tracks = get_wanted_tracks(love_me, types)
    print(tracks)
test_get_wanted_tracks()
```

```python
def filter_song(song, types, min_length):
    tracks = get_wanted_tracks(song, types)
    if not tracks or len(tracks) != len(types):
        return None
    mid = mido.MidiFile()
    for track in tracks:
        if track == -1 or len(track) < min_length:
            return None
        mid.tracks.append(track)
    return mid
```

```python
def test_filter_song():
    love_me = mido.MidiFile('../../songs/small_db/Love_Me_Do.mid')
    types = ["Organ", "Bass"]
    filtered_song = filter_song(love_me, types, 50)
    print(filtered_song)

test_filter_song()
```

```python
def filter_songs(songs, instruments_types, min_length):
    filtered_songs = []
    for song in songs:
        filtered_song = filter_song(song, instruments_types, min_length)
        if filtered_song:
            filtered_songs.append(filtered_song)
    return filtered_songs
```

```python
instruments_types = ["Guitar", "Bass"]
filtered_songs = filter_songs(songs, instruments_types, 200)
```

```python
def clean_songs(songs):
    # returns songs with only note_on's
    clean_songs = []
    for old_song in songs:
        new_song = mido.MidiFile()
        for old_track in old_song.tracks:
            new_track = mido.MidiTrack()
            for msg in old_track:
                if msg.type == "note_on":
                    new_track.append(msg)
            new_song.tracks.append(new_track)
        clean_songs.append(new_song)
    return clean_songs
```
